```python
# df = dataset_utils.df_combined(data)
# df = pd.DataFrame(data)
df = dataset_utils.df_combined(data)
min_length = 40 #set the minimum size of point clouds
df.shape
```

```python
#only keep frames with at least min_length data points 
df = df[df["IWR1843_Ori"].apply(lambda x: len(x) >= min_length)]
df
```

```python
# Split the DataFrame into train and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Parameters used in saving the file

experiment_name = "standing"  # Replace with your actual experiment name
file_label = "auto_RadarSeq"
file_time = "Jul-24-00-50-36"  # Replace with the actual time string when the file was saved
train_dir = f"pointnet_data/{category}/train"  # Replace with the actual directory path
test_dir = f"pointnet_data/{category}/test"
```

```python
try:
    os.makedirs(train_dir)
    os.makedirs(test_dir)
except:
    pass
```

```python
for i, row in train_df.iterrows():
    print(row)
    file_name = f"{experiment_name}_{file_label}_{file_time}_{i}"
    file_path = os.path.join(train_dir, file_name)
    row.to_pickle(file_path)
    
```

```python
for i, row in test_df.iterrows():
    print(row)
    file_name = f"{experiment_name}_{file_label}_{file_time}_{i}"
    file_path = os.path.join(test_dir, file_name)
    row.to_pickle(file_path)
    
```
